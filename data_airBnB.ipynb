# Exploration multivariée

### 1.) Quartier, prix et année de construction

```python
df_10yrs = df[(df.loc[:,'construction_year'] >2014)
             & (df.loc[:,'construction_year'] <= 2022)]
```

```python
plt.figure(figsize=(6,6))
sb.catplot(x='neighbourhood_group',y='price',data=df_10yrs,hue='construction_year',kind='point',
          col='construction_year', col_wrap=3)
```

Il semble qu'avec une année de construction plus récente, 2018 et plus tard, le prix moyen des annonces dans le Queens, Staten Island et le Bronx peut commander un prix plus élevé que Brooklyn et Manhattan. Enfin, quelque chose qui semble intuitif.

### 2.) Availability, Price, and Neighborhood

```python
xticks = np.arange(0,450,50)
plt.hist(x='availability_365',data=df, bins= 60)
plt.xlim(0,600)
plt.xticks(xticks, rotation=30);
```

Pour être honnête, je ne suis toujours pas sûr à 100% de ce que représente la fonction " availability_365 ". Mon hypothèse la plus probable, basée sur la définition interne d'Airbnb, est qu'il s'agit d'une représentation de la disponibilité des annonces 365 jours dans le futur. Très difficile à discerner étant donné qu'il y avait des valeurs négatives à l'origine et des valeurs maximales jusqu'à 3677. Je pars du principe que plus le nombre de disponibilité est faible, plus l'annonce est réservée fréquemment, et vice-versa. 

```python
print(df.availability_365.describe())
print(df.availability_365.max())
```
